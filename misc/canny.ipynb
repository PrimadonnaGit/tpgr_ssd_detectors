```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def plot_box(box, box_format='xywh', color='r', linewidth=1):
    if box_format == 'xywh':
        xmin, ymin, w, h = box
        xmax, ymax = xmin + w, ymin + h
        xy_rec = np.array([[xmin, ymin], [xmax, ymin], [xmax, ymax], [xmin, ymax]])
    elif box_format == 'xyxy':
        xmin, ymin, xmax, ymax = box
        xy_rec = np.array([[xmin, ymin], [xmax, ymin], [xmax, ymax], [xmin, ymax]])
    else: # polygon
        pass
    ax = plt.gca()
    ax.add_patch(plt.Polygon(xy_rec, fill=False, edgecolor=color, linewidth=linewidth))
```

```python
img = cv2.imread('../data/RoboTT/8anno/BoxLotDate_MP-42790-001-01-LV-001(2).jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (7,7), 0)
edges = cv2.Canny(blur, 20, 80)
box = cv2.boundingRect(cv2.findNonZero(edges))

plt.imshow(img/255)
plot_box(box)
plt.show()

plt.imshow(edges)
plot_box(box)
plt.show()
```
